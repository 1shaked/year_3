```python
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import math
```

```python
e = np.e
c = 1/(1.5 * e - 1)
A = c* (e - 1)
A, c
```

```python
def get_random_point(u):
    if u <= A:
        return np.log( (u/c) + 1)
    return (3 - np.log(e - 2 * ( u - A) /c )) / 2
```

```python
def generate_points(n):
    points = []
    for i in range(n):
        u = np.random.uniform(0, 1)
        points.append(get_random_point(u))
    return points

```

```python
points = generate_points(1_500)
# generate histogram
fig = px.histogram(points, nbins=50,  title="Histogram of Random Points")
fig.update_traces(marker=dict(line=dict(width=1, color='black')))
fig.update_layout(
    xaxis_title="Value",
    yaxis_title="Frequency",
    title_x=0.5,
    title_y=0.95,
    font=dict(size=14),
    width=800,
    height=600
)
fig.update_xaxes(
    tickvals=np.arange(0, 3.5, 0.5),
    ticktext=[f"{i:.1f}" for i in np.arange(0, 3.5, 0.5)],
    title_standoff=10,
)
fig.update_yaxes(
    tickvals=np.arange(0, 200, 20),
    ticktext=[f"{i}" for i in np.arange(0, 200, 20)],
    title_standoff=10,
)
fig.show()
```
