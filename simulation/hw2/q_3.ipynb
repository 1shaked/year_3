```python
import statistics

import random
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from tqdm import tqdm

```

```python
from scipy.stats import binom

# Parameters
n = 104       # number of trials
p = 19/20     # probability of success
k = 100        # number of successes

# 1. Probability of exactly k successes (P(X = k))
prob = binom.pmf(k, n, p)
print(f"P(X = {k}) = {prob}")

# 2. Cumulative probability (P(X ≤ k))
cum_prob = binom.cdf(k, n, p)
print(f"P(X ≤ {k}) = {cum_prob}")

# 3. Probability of more than k successes (P(X > k))
greater_prob = 1 - binom.cdf(k, n, p)
print(f"P(X > {k}) = {greater_prob}")
```

```python
n = 5_000
p = 0.95
cost = 80
available_rooms = 100
min_rooms_size = 100
max_rooms_size = 110
rev_track_list = []
for rooms_num in range(min_rooms_size, max_rooms_size + 1):
    for i in range(n):
        rev_track = []
        total_persons = 0
        for person in range(rooms_num):    
            v = random.randint(0, 20)
            if v <= 19:
                total_persons += 1
        rev = cost * min(total_persons, available_rooms) - 1.6 * cost * (max(0, total_persons - available_rooms))
        rev_track.append(rev)
    avg_track = statistics.mean(rev_track)
    rev_track_list.append(avg_track)


# plot the rev track list
plt.plot(rev_track_list)
plt.ylabel('Revenue')
plt.xlabel('Number of rooms')

max(rev_track_list)
# find the index of the max value
max_index = rev_track_list.index(max(rev_track_list))
print(f"Max revenue is {max(rev_track_list)} for {min_rooms_size + max_index} rooms")
```
