```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
df = pd.read_csv('audio_compression_results.csv')
df.head()
```

```python
avg_compression_size = {
    "RLE": df["rle_compressed_size"].mean(),
    "Huffman": df["huffman_compressed_size"].mean(),
    "LZW": df["lzw_compressed_size"].mean(),
}
avg_compression_time = {
    "RLE": df["rle_time"].mean(),
    "Huffman": df["huffman_time"].mean(),
    "LZW": df["lzw_time"].mean(),
}
```

```python
avg_compression_size
```

```python

# # Plot 1: Average Compression Size
# plt.bar(avg_compression_size.keys(), avg_compression_size.values())
# plt.title("Average Compression Size per Algorithm")
# plt.ylabel("Size (Bytes)")
# plt.xlabel("Algorithm")
# plt.show()
# 1. Average Compression Size
fig1 = go.Figure()
fig1.add_trace(go.Bar(x=list(avg_compression_size.keys()), y=list(avg_compression_size.values()), name="Avg Size"))
fig1.update_layout(
    title="Average Compression Size per Algorithm",
    xaxis_title="Algorithm",
    yaxis_title="Size (Bytes)",
)
fig1.show()
```

```python
# Plot 2: Average Compression Time
plt.bar(avg_compression_time.keys(), avg_compression_time.values())
plt.title("Average Compression Time per Algorithm")
plt.ylabel("Time (Seconds)")
plt.xlabel("Algorithm")
plt.show()
```

```python
# Plot 3: Distribution of Compressed Sizes
plt.boxplot(
    [df["rle_compressed_size"], df["huffman_compressed_size"], df["lzw_compressed_size"]],
    labels=["RLE", "Huffman", "LZW"],
)
plt.title("Distribution of Compression Sizes")
plt.ylabel("Size (Bytes)")
plt.xlabel("Algorithm")
plt.show()
```

```python
# Plot 4: Lowest Compression Size for Each Algorithm
min_compression_size = {
    "RLE": df["rle_compressed_size"].min(),
    "Huffman": df["huffman_compressed_size"].min(),
    "LZW": df["lzw_compressed_size"].min(),
}
plt.bar(min_compression_size.keys(), min_compression_size.values(), color=["red", "green", "blue"])
plt.title("Lowest Compression Size per Algorithm")
plt.ylabel("Size (Bytes)")
plt.xlabel("Algorithm")
plt.show()
```
