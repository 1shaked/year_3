```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df = pd.read_csv('train.csv')

df.head()
```

```python
labels = list(sorted(df.label.unique()))
labels
```

```python
EXAMPLE_NUMBER = 4
fig , axes = plt.subplots(len(labels), EXAMPLE_NUMBER, figsize=(10, 20))
fig.tight_layout(pad=3.0)

# Loop through each label and for each one plot 4 images
for label_number in labels:
    for i in range(EXAMPLE_NUMBER):
        ax = axes[label_number, i]
        ax.imshow(df[df.label == label_number].iloc[i, 1:].values.reshape(28, 28), cmap='gray')
        ax.set_title(f'Label: {label_number}')
        ax.axis('off')

plt.show()
```
