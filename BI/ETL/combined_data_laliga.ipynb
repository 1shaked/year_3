```python
df['home_shots'] = df['Home Team Off Target Shots'] + df['Home Team On Target Shots']

# Example column mapping
column_mapping = {
    'Home Team': 'home_team',
    'Away Team': 'away_team',
    'Score': 'result_full',
    'Half Time Score': 'result_ht',
    'Home Team Possession %': 'home_possession',
    'Away Team Possession %': 'away_possession',
    'home_shots': 'home_shots',
    # 'Home Team Off Target Shots': 'home_shots',
    'Home Team On Target Shots': 'home_shots_on_target',
    'Home Team Total Shots': 'home_shots',
    'Home Team Blocked Shots': 'home_tackles',
    'Home Team Corners': 'home_corners',
    'Home Team Clearances': 'home_clearances',
    'Home Team Fouls': 'home_fouls_conceded',
    'Home Team Yellow Cards': 'home_yellow_cards',
    'Home Team Red Cards': 'home_red_cards',
    'Away Team Off Target Shots': 'away_shots',
    'Away Team On Target Shots': 'away_shots_on_target',
    'Away Team Total Shots': 'away_shots',
    'Away Team Blocked Shots': 'away_tackles',
    'Away Team Corners': 'away_corners',
    'Away Team Clearances': 'away_clearances',
    'Away Team Fouls': 'away_fouls_conceded',
    'Away Team Yellow Cards': 'away_yellow_cards',
    'Away Team Red Cards': 'away_red_cards',
    'Home Team Goals Scored': 'goal_home_ft',
    'Away Team Goals Scored': 'goal_away_ft',
    'year': 'season',
    # Add more mappings as needed
}

# Rename columns in df_new to match the current database
df = df.rename(columns=column_mapping)
# Drop any columns in df_new that are not needed
columns_to_keep = column_mapping.values()
df_new_adjusted = df[list(columns_to_keep)]
```

```python
with open('../data/data_dictionary.txt', 'r') as f:
    all_fields = f.readlines()
    all_fields = map(lambda x: x.split('\t')[0], all_fields)
    all_fields = list(all_fields)
print(all_fields)
```

```python
# get the differences between the columns in the data and the columns in the data dictionary
diff = set(all_fields) - set(df.columns)
for i, v in enumerate(diff):
    if '_avg_' in v or '_acum_' in v:
        continue
    print(i, v)
```

```python
# heatmap between the result and the other columns
# shooting accuracy (shots on target / total shots)
# the relation between the first round to the second round ,
df['sg_match_ft']  = df['goal_home_ft'] - df['goal_away_ft']
df.to_csv('../data/combined_data_laliga_tranform.csv', index=False)
```
