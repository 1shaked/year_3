```python
import pandas as pd
import statsmodels.api as sm
# Table data provided by the user
data_full = {
    'Combination': ['(1)', 'a', 'b', 'ab', 'c', 'ac', 'bc', 'abc'],
    'Replicate I': [325, 211, 532, 392, 420, 406, 560, 354],
    'Replicate II': [435, 311, 501, 419, 473, 377, 543, 348],
    'Replicate III': [356, 220, 504, 401, 442, 387, 557, 352],
    'Replicate IV': [415, 295, 514, 405, 451, 381, 547, 349]
}

# Create a DataFrame from the table
df_full = pd.DataFrame(data_full)

# Calculate the average response for each combination
df_full['Y'] = df_full[['Replicate I', 'Replicate II', 'Replicate III', 'Replicate IV']].mean(axis=1)

# Add encoded factor levels for x1, x2, x3 based on the combination
factor_encoding = {
    '(1)': [-1, -1, -1],
    'a': [1, -1, -1],
    'b': [-1, 1, -1],
    'ab': [1, 1, -1],
    'c': [-1, -1, 1],
    'ac': [1, -1, 1],
    'bc': [-1, 1, 1],
    'abc': [1, 1, 1]
}

# Map the factors to the DataFrame
df_full[['x1', 'x2', 'x3']] = df_full['Combination'].map(factor_encoding).apply(pd.Series)

# Add interaction terms
df_full['x1x2'] = df_full['x1'] * df_full['x2']
df_full['x1x3'] = df_full['x1'] * df_full['x3']
df_full['x2x3'] = df_full['x2'] * df_full['x3']
df_full['x1x2x3'] = df_full['x1'] * df_full['x2'] * df_full['x3']

# Define independent variables (including interaction terms)
X_full = df_full[['x1', 'x2', 'x3', 'x1x2', 'x1x3', 'x2x3', 'x1x2x3']]

# Add a constant for the intercept
X_full = sm.add_constant(X_full)

# Define the dependent variable
y_full = df_full['Y']

# Fit the model
model_full = sm.OLS(y_full, X_full).fit()

# Display the model summary
# import ace_tools as tools; tools.display_dataframe_to_user(name="Factorial Design Data Table", dataframe=df_full)
model_full.summary()
```
