```python
import matplotlib.pyplot as plt
import math as m
import numpy as np
```

```python
probabilities = [x/100 for x in range(0, 17)]
```

```python
n=200
c=7
```

```python
def passion(lambda_avg, k):
    return m.exp(-lambda_avg)*((lambda_avg)**k)/m.factorial(k)

def cumulative_passion(lambda_avg, k):
    return sum([passion(lambda_avg, x) for x in range(0, k+1)])
```

```python
AOQ = []
for i in probabilities:
    p_a = cumulative_passion(n*i, c)
    AOQ.append(p_a * i)
```

```python
# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(probabilities, AOQ, color='blue', linewidth=2, label='AOQ Curve')

# Add labels and title
plt.title("Average Outgoing Quality (AOQ)", fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Proportion Defective (p)", fontsize=12, color='darkblue')
plt.ylabel("Average Outgoing Quality (AOQ)", fontsize=12, color='darkblue')

# Customize the grid
plt.grid(visible=True, linestyle='--', linewidth=0.5, alpha=0.7)

# Customize the axes
plt.xticks(fontsize=10, color='darkblue')
plt.yticks(fontsize=10, color='darkblue')

# Add legend
plt.legend(fontsize=10, loc='upper right', frameon=True, shadow=True)

# Highlight AOQL (maximum AOQ)
max_AOQ = max(AOQ)
max_index = np.argmax(AOQ)
plt.scatter(probabilities[max_index], max_AOQ, color='red', label=f"AOQL ({max_AOQ:.4f})")
plt.annotate(f"AOQL\n({probabilities[max_index]:.2f}, {max_AOQ:.4f})",
             xy=(probabilities[max_index], max_AOQ),
             xytext=(probabilities[max_index] + 0.01, max_AOQ + 0.005),
             arrowprops=dict(facecolor='red', arrowstyle="->"),
             fontsize=10, color='red')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
