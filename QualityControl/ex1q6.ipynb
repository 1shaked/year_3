```python
import matplotlib.pyplot as plt
import statistics as stats
import pandas as pd
```

```python
ucl = 0.365
lcl = 0.35
sigma = 0.002
avg = (ucl + lcl) / 2
def cpk(lcl , ucl , sigma , x):
    return min((ucl - x) / (3 * sigma) , (x - lcl) / (3 * sigma)) 

def cp(lcl , ucl , sigma):
    return (ucl - lcl) / (6 * sigma)
```

```python
total = []
for i in range(0 , 101):
    v = 4 * sigma * i / 100
    avg_top = avg + v
    avg_bottom = avg - v
    total.append({
        'x' : avg_top,
        'cpk' : cpk(lcl , ucl , sigma , avg_top),
        'cp' : cp(lcl , ucl , sigma)
    })
    total.append({
        'x' : avg_bottom,
        'cpk' : cpk(lcl , ucl , sigma , avg_bottom),
        'cp' : cp(lcl , ucl , sigma)
    })
```

```python
total
```

```python
df = pd.DataFrame(total)
df
```

```python
plt.plot(df['x'] , df['cpk'] , label = 'cpk')
plt.plot(df['x'] , df['cp'] , label = 'cp', linestyle = '--')
plt.show()
```
