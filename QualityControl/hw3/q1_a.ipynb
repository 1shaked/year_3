```python

import math as m
import matplotlib.pyplot as plt
```

```python
def passion(n, k , lambda_avg=None):
    if lambda_avg is None:
        lambda_avg = n
    return (lambda_avg**k * m.exp(-lambda_avg)) / m.factorial(k)

def passion_cumulative(n, k , lambda_avg=None):
    p = 0
    for i in range(k):
        p += passion(n, i, lambda_avg)
    return p
```

```python
probabilities = [x/8 for x in range(0, 72)]
# print probabilities
```

```python
c = 2
n = 100

probabilities_accept = [passion_cumulative(n, c , x) for x in probabilities]
```

```python

# plt.style.use('dark_background')
# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(probabilities, probabilities_accept, marker='o', label='Acceptance Probability')

# Add labels and title
plt.title('OC Curve', fontsize=16, fontweight='bold')
plt.xlabel('Defective Probability (p)', fontsize=12)
plt.ylabel('Acceptance Probability (Pa)', fontsize=12)

# Add grid
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend
plt.legend(fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```
