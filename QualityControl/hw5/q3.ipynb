```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
```

```python
df = pd.read_excel('q3.xlsx')
```

```python
data = df.values.flatten().tolist()
```

```python

# Create Q-Q plot for a normal distribution
stats.probplot(data, dist="norm", plot=plt)

# Add a title and show the plot
plt.title("Q-Q Plot")
plt.show()
```

```python
df.plot(kind='line', marker='o', figsize=(10, 6))

# Add labels and title
plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Line Plot for Each Column in DataFrame")
plt.legend(title="Columns")

# Show the plot
plt.show()
```

```python
# Sort data points in each column
sorted_df = df.apply(sorted, axis=0)  # Sort each column individually

# Plot each column as a scatter line plot
plt.figure(figsize=(10, 6))  # Set figure size

for col in sorted_df.columns:
    plt.plot(sorted_df.index, sorted_df[col], marker='o', linestyle='-', label=col)

# Add labels, legend, and title
plt.xlabel("Sorted Index")
plt.ylabel("Values")
plt.title("Scatter Line Plot for Each Sorted Column")
plt.legend(title="Columns")

# Show the plot
plt.show()
```
